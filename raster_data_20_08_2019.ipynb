```python
from PIL import Image
import numpy as np
import pandas as pd 
import os
from pandas import ExcelWriter
from pandas import ExcelFile


def read_tif(address):
    return np.array(Image.open(address))
```

```python
def get_file_name(directory):
    file_name_list = []
    for file in os.listdir(directory):
        if file.endswith(".tif"):
            file_name_list.append(file)
    return file_name_list
```

```python
file_name = get_file_name(os.path.abspath("C:/Users/klin0002/Downloads/raster data"))
# file_name.remove('SummerLandsat75_300_900m.tif')
file_name
```

```python
file_name.remove('hydro500xwi.tif')
file_name.remove('SummerLandsat75_300_900m.tif')
file_name.remove('SummerPre1750Landsat75_300_900m.tif')
```

```python
df = pd.read_excel('aaa.xls')
valuelist= []

for i in range(len(file_name)):
    value_one_file=[]
    imarray= read_tif(file_name[i])
    unit_lati = abs(-32.8082-(-39.1763))/len(imarray)
    unit_log = abs(138.871-150.685)/len(imarray[0])
    for j in range(len(df['TAXON_TYPE'])):
        index_lati=round(abs((-32.8082-df["LATITUDEDD_NUM"][j])/unit_lati))
        index_log = round(abs((df["LONGITUDEDD_NUM"][j]-138.871)/unit_log))
        value_one_file.append(imarray[int(index_lati)][int(index_log)])
    valuelist.append(value_one_file)
a = pd.DataFrame(valuelist)   
a= a.transpose()
a.columns=file_name
a
```
